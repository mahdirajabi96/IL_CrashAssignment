```python
import pandas as pd
import numpy as np
import time
```

```python
LocIDs_DF = pd.read_csv('tcds_list.csv',skiprows=1,low_memory=False)
LocIDs_DF = LocIDs_DF.rename(columns={'Latest Date':'LatestDate','Loc ID':'LocID'})
LocIDs_DF['LatestDate'] = pd.to_datetime(LocIDs_DF['LatestDate'])
LocIDs_DF.index = LocIDs_DF['LocID']
LocIDs_DF.shape
```

```python
HVL_DF = LocIDs_DF.loc[LocIDs_DF.Dir.isin([str(i) for i in range(1,8)])]
idx1 = pd.Index(HVL_DF.LocID.apply(lambda x:x.split('_')[0]))
idx2 = pd.Index(HVL_DF.LocID.apply(lambda x:x.split('_')[2]))
idx3 = pd.Index(HVL_DF.LocID.apply(lambda x:x.split('_')[1]))
HVL_DF.index = pd.MultiIndex.from_arrays([idx1,idx3,idx2])
HVL_DF.shape
```

```python
hourlyurl = 'http://idot.ms2soft.com/tcds/TCDS_HourlyVolByLane.aspx?&a={}&local_id={}&st_date={}'.format
HVL_DF.loc[:,'LocID1'] = HVL_DF.LocID.apply(lambda x:x.split('_')[0])
s = HVL_DF.groupby('LocID1')['LatestDate'].first()
H_DF = pd.DataFrame()
for locid in s.index:
    time.sleep(1)
    try:
        d = s.loc[locid]
        url = hourlyurl(44,locid.replace(' ','%20'),d.strftime('%Y-%m-%d'))
        print(url)
        PD = pd.read_html(url)
        df = PD[0]
        df.columns = df.loc[0]
        df = df.loc[1:24]
        df = df.rename(columns={np.nan:'Time'})
        df = df.drop('Time',axis=1)
        CL = ['{}_{}_{}'.format(locid,i[0],i[1]) for i in pd.MultiIndex.from_product([range(1,7),['NB','SB','EB','WB']]).tolist()]
        CL = [cl for cl in CL if cl in df.columns]
        df = df[CL]
        df.columns = pd.MultiIndex.from_tuples([tuple(c.split('_')[1:]) for c in df.columns])
        idx1 = [pd.datetime(d.year,d.month,d.day,i,0) for i in range(0,24) ]
        df['LocID'] = locid
        df['DATE'] = idx1
        df.index = pd.MultiIndex.from_arrays([pd.Index([locid]*24),idx1])
        H_DF = pd.concat([H_DF,df])
    except:
        print('failed')
H_DF
```
