```python
# Fix Index
# In 2007 and 2008 crash data there are 19 and 99 duplicated case numbers accordingly. These are fixed here by
#starting new ids from 20XX99900000
for y in range(2007,2009):
    df  = pd.read_csv(SourceDir+'\\{}_Crash.csv'.format(y),index_col=0,encoding='latin-1',low_memory=False)
    vdf = pd.read_csv(SourceDir+'\\{}_Vehicle.csv'.format(y),index_col=0,encoding='latin-1',low_memory=False)
    odf = pd.read_csv(SourceDir+'\\{}_Person.csv'.format(y),index_col=0,encoding='latin-1',low_memory=False)

    if y==2016:
        df.CaseID_CH2M = df.CaseNum
    df['ID_XY'] = ['{}_{}_{}'.format(i,x,y) for i,x,y in zip(df.CaseID_CH2M,df.X_COORD,df.Y_COORD)]
    df['New_ID'] = df.CaseID_CH2M

    vdf['ID_XY'] = ['{}_{}_{}'.format(i,x,y) for i,x,y in zip(vdf.Case_number_CH2M,vdf.X_Coord,vdf.Y_Coord)]
    vdf['New_ID'] = vdf.Case_number_CH2M

    odf['ID_XY'] = ['{}_{}_{}'.format(i,x,y) for i,x,y in zip(odf.CaseNum_CH2M,odf.X_Coord,odf.Y_Coord)]
    odf['New_ID'] = odf.CaseNum_CH2M

    dup_idx = list(set([i for i,r in df.CaseID_CH2M.value_counts().iteritems() if r>1]))
    idf = pd.DataFrame(columns=['ID','ID_XY'])
    i = 0
    for idx in dup_idx:
        for n,m in df[df.CaseID_CH2M==idx].iterrows():
            i += 1
            idf.loc[i] = [idx,m.ID_XY]
    n = idf.loc[idf.ID.duplicated()].shape[0]
    idf.loc[~idf.ID.duplicated(),'New_ID'] = idf.loc[~idf.ID.duplicated(),'ID']
    m = int('{}99900000'.format(y))
    idf.loc[idf.ID.duplicated(),'New_ID'] = list(range(m,m+n))

    df.index = df.ID_XY
    df.loc[idf.ID_XY,'New_ID'] = list(idf.New_ID)

    vdf.index = vdf.ID_XY
    for idxy,newid in zip(idf.ID_XY,idf.New_ID):
        if idxy in vdf.index:
            vdf.loc[idxy,'New_ID'] = newid

    odf.index = odf.ID_XY
    for idxy,newid in zip(idf.ID_XY,idf.New_ID):
        if idxy in odf.index:
            odf.loc[idxy,'New_ID'] = newid

    print(y,
          len(dup_idx),
          df[df.CaseID_CH2M!=df.New_ID].shape[0],
          vdf[vdf.Case_number_CH2M!=vdf.New_ID].shape[0],
          odf[odf.CaseNum_CH2M!=odf.New_ID].shape[0])
    df.Casenum = df.New_ID
    vdf.Case_number = vdf.New_ID
    odf.CaseNum = odf.New_ID
    df.to_csv(SourceDir+'\\{}_Crash_FixIndex.csv'.format(y))
    vdf.to_csv(SourceDir+'\\{}_Vehicle_FixIndex.csv'.format(y))
    odf.to_csv(SourceDir+'\\{}_Person_FixIndex.csv'.format(y))
```

```python

df = Occ_DF[Occ_DF.INJ_SEV==4]
df['UID'] = list(df.index.droplevel([0,2]))
def Ex_Unit(u):
    if u==1:
        return('Unit1')
    else:
        return('OtherUnits')
def Ex_PT(u):
    if u==1:
        return('Driver')
    elif u in [2,3]:
        return('Front Passenger')
    elif u==0:
        return('Unknown')
    else:
        return('BackSeat Passengers')
df['Unit'] = [Ex_Unit(u) for u in df.UID]
df['PersonType'] = [Ex_PT(p) for p in df.SEAT_POS]
def Ex_SE(s):
    if s in [1,3,5,7,8]:
        return('Improper/Not Used')
    elif s in [9]:
        return('Unknown')
    else:
        return('Used')
df['SafetyEq'] = [Ex_SE(s) for s in df.SAF_EQI]
```

```python
def ExtractINV(s):
    s = str(s)
    l = s.split('_')
    if len(l)==2:
        inv = str(l[0])
        mp = float(l[1])
        bmp = mp-250.0/5280.0
        emp = mp+250.0/5280.0
        return(inv)
    if len(l)==3:
        inv = str(l[0])
        bmp = float(l[1])
        emp = float(l[2])
        return(inv)
def ExtractBMP(s):
    s = str(s)
    l = s.split('_')
    if len(l)==2:
        inv = str(l[0])
        mp = float(l[1])
        bmp = mp-250.0/5280.0
        emp = mp+250.0/5280.0
        return(bmp)
    if len(l)==3:
        inv = str(l[0])
        bmp = float(l[1])
        emp = float(l[2])
        return(bmp)
def ExtractEMP(s):
    s = str(s)
    l = s.split('_')
    if len(l)==2:
        inv = str(l[0])
        mp = float(l[1])
        bmp = mp-250.0/5280.0
        emp = mp+250.0/5280.0
        return(emp)
    if len(l)==3:
        inv = str(l[0])
        bmp = float(l[1])
        emp = float(l[2])
        return(emp)
df['INV1'] = [ExtractINV(s) for s in df.INTX_SEG_ID]
df['BMP1'] = [ExtractBMP(s) for s in df.INTX_SEG_ID]
df['EMP1'] = [ExtractEMP(s) for s in df.INTX_SEG_ID]
```
