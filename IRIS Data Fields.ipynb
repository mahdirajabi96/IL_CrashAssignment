```python
IRIS_Shp   = {y:os.path.join(IRISPath,'HWY'+str(y)+'.shp') for y in Years}
df = pd.DataFrame(columns=Years)
for y in Years:
    LF = [F.name for F in arcpy.ListFields(IRIS_Shp[y])]
    for f in LF:
        df.set_value(f,y,1)
df = df.fillna(' ')
df = df.sort_index()
HTML(df.to_html())
```
